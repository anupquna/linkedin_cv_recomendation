```python
def create_data2(filename):
    #ilename1=filename.replace('_', ' ')
    filename1=filename[8:]
    filename1=filename1.replace('.docx', '')
    my_text = docx2txt.process(filename)
    my_df = pd.DataFrame()
    my_df = pd.DataFrame(data=[filename1],columns=['Name'])
    # using split() 
    # to extract words from string 
    res = my_text.split('\n\n') 
    kk=''
    cnt=0
    for t in res:
        cnt=cnt+1
        if t=='Contact':
            kk=''
            cnt2=cnt
            for t in res:  
                kk=kk+res[cnt2]
                cnt2=cnt2+1
                if res[cnt2]=='':
                    my_df['Contact'] =kk
                    break 
        if "Top Skills" in t:
            cnt2=cnt
            kk=''
            for t in res:
                kk=kk+res[cnt2]
                cnt2=cnt2+1
                if res[cnt2]=='':
                    my_df['Top_Skills'] =kk
                    break
        if t=='Languages':
            cnt2=cnt
            kk=''
            for t in res:      
                kk=kk+res[cnt2]
                cnt2=cnt2+1
                if res[cnt2]=='':
                    my_df['Languages'] =kk
                    break
        if t=='Certifications':
            cnt2=cnt
            kk=''
            for t in res:      
                kk=kk+res[cnt2]
                cnt2=cnt2+1
                if res[cnt2]=='':
                    my_df['Certifications'] =kk
                    break
        if t==filename1:
            cnt2=cnt
            kk=''
            for t in res:      
                kk=kk+res[cnt2]
                cnt2=cnt2+1
                if res[cnt2]=='':
                    my_df['Designation'] =kk
                    break  
        if t=='Summary':
            cnt2=cnt
            kk=''
            for t in res:
                kk=kk+res[cnt2]
                cnt2=cnt2+1
                if res[cnt2]=='':
                    my_df['Summary'] =kk
                    break
        if t=='Experience':
            cnt2=cnt
            kk=''
            for t in res:           
                kk=kk+res[cnt2]
                cnt2=cnt2+1
                if cnt2+1 > len(res):
                    my_df['Experience'] =kk
                    break
                else:
                    if res[cnt2]=='Education':
                        my_df['Experience'] =kk
                        break
        if t=='Education':
            cnt2=cnt
            kk=''
            for t in res:      
                kk=kk+res[cnt2]
                cnt2=cnt2+1
                if res[cnt2]=='':
                    my_df['Education'] =kk
                    break
     
        if 'Contact' in my_df.columns:
            str='ok'
        else:
            my_df['Contact'] ='' 
        if 'Top_Skills' in my_df.columns:
            str='ok'
        else:
            my_df['Top_Skills'] ='No Skill'
        if 'Languages' in my_df.columns:
            str='ok'
        else:
            my_df['Languages'] =''
        if 'Certifications' in my_df.columns:
            str='ok'
        else:
            my_df['Certifications'] =''
        if 'Summary' in my_df.columns:
            str='ok'
        else:
            my_df['Summary'] =''
        if 'Designation' in my_df.columns:
            str='ok'
        else:
            my_df['Designation'] =''
        if 'Experience' in my_df.columns:
            str='ok'
        else:
            my_df['Experience'] =''
        if 'Education' in my_df.columns:
            str='ok'
        else:
            my_df['Education'] =''
            
    my_df1=my_df.values.tolist()
    return my_df1,filename1
```

```python
import re

def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
# Combining all the above statemennts 
from tqdm import tqdm
preprocessed_Summary = []
# tqdm is for printing the status bar
for sentance in tqdm(ddfff['Summary'].values):
    sent = decontracted(sentance)
    sent = sent.replace('\\r', ' ')
    sent = sent.replace('\\"', ' ')
    sent = sent.replace('\\n', ' ')
    sent = re.sub('[^A-Za-z0-9]+', ' ', sent)
    sent = ' '.join(e for e in sent.split() if e not in stopwords)
    preprocessed_Summary.append(sent.lower().strip())
```

```python
# Combining all the above statemennts 
from tqdm import tqdm
preprocessed_Top_Skills = []
# tqdm is for printing the status bar
for sentance in tqdm(ddfff['Top_Skills'].values):
    sent = decontracted(sentance)
    sent = sent.replace('\\r', ' ')
    sent = sent.replace('\\"', ' ')
    sent = sent.replace('\\n', ' ')
    sent = re.sub('[^A-Za-z0-9]+', ' ', sent)
    sent = ' '.join(e for e in sent.split() if e not in stopwords)
    preprocessed_Top_Skills.append(sent.lower().strip())
```

```python
# Combining all the above statemennts 
from tqdm import tqdm
preprocessed_experiance = []
# tqdm is for printing the status bar
for sentance in tqdm(ddfff['Experience'].values):
    sent = decontracted(sentance)
    sent = sent.replace('\\r', ' ')
    sent = sent.replace('\\"', ' ')
    sent = sent.replace('\\n', ' ')
    sent = re.sub('[^A-Za-z0-9]+', ' ', sent)
    sent = ' '.join(e for e in sent.split() if e not in stopwords)
    preprocessed_experiance.append(sent.lower().strip())
```
